```python
import pandas as pd 
```

```python
Amazon_books = pd.read_csv('../../data/raw/Amazon/books_data.csv')
```

```python
Amazon_reviews = pd.read_csv('../../data/raw/Amazon/Books_rating.csv')
Amazon_reviews.head()
```

```python
Amazon_reviews = Amazon_reviews.rename(columns={ "Id" : "ISBN" ,
                                 "review/score" : "rating" , 
                                 "review/text" : "review"
                               })
```

```python
Amazon_reviews.head()
```

```python
Amazon_books = Amazon_books.merge(Amazon_reviews , how = 'left' , on = "Title")
```

```python
Amazon_reviews = Amazon_reviews[['ISBN' , 'User_id' , 'rating' , 'review']].dropna()
Amazon_reviews.head()
```

```python
Amazon_reviews.isnull().sum()
```

```python
Amazon_reviews['ISBN'].values
```

```python
Amazon_reviews.to_csv('../../data/processed/amazon_reviews_p.csv',index=False)
```

```python
Amazon_books = Amazon_books.drop_duplicates(subset=["ISBN"]).reset_index(drop=True)
```

```python
Amazon_books
```

```python
Amazon_books.rename(columns={ 
                              'Title':'Book-Title' , 
                              'description' : 'Description' , 
                              'authors':'Book-Author',
                              'publisher' : 'Publisher' , 
                              'publishedDate' : 'Year-Of-Publication' ,
                              'image' : 'Image' ,
                              'categories' : 'Categories'
                             },inplace=True
                             )
Amazon_books = Amazon_books[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher', 'Description', 'Categories' , 'Image']]
Amazon_books.head()
```

```python
import re
Amazon_books['Year-Of-Publication'] = Amazon_books['Year-Of-Publication'].apply(lambda x: re.findall(r'\d{4}', str(x))[0] if re.findall(r'\d{4}', str(x)) else None)
Amazon_books['Year-Of-Publication'] = Amazon_books['Year-Of-Publication'].astype(float)
Amazon_books['Year-Of-Publication'].value_counts()
```

```python
Amazon_books['word_count_per_desc'] = Amazon_books['Description'].str.split().str.len()
```

```python
Amazon_books = Amazon_books[Amazon_books['word_count_per_desc'] > 5].reset_index(drop=True)
Amazon_books.head()
```

```python
Amazon_books.isnull().sum()
```

```python
Amazon_books.dropna(inplace=True)
```

```python
Amazon_books.isnull().sum()
```

```python
Amazon_books.head()
```

```python
Amazon_books.drop(columns=["word_count_per_desc"] , inplace=True)
```

```python
Amazon_books.reset_index(drop=True)
Amazon_books.head()
```
