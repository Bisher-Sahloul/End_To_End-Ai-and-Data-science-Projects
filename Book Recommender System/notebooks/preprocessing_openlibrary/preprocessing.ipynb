```python
import pandas as pd 
```

```python
df = pd.read_csv('../../data/raw/Openlibrary/openlibrary_books.csv')
df.head()
```

```python
df.rename(columns={
                                     'Subject':'Categories',
                                     'Image-URL-M':'Image',
                                  }, inplace=True)
df = df[['ISBN', 'Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher' , 'Description' , 'Categories', 'Image']]
df.head()
```

```python
df["ISBN"] = df["ISBN"].str.split("\n")
df["ISBN"] = df["ISBN"].str[0].str.strip().str.strip(',')
```

```python
# Aceepting only ISBN 10 
df = df[df["ISBN"].apply(lambda x : len(x) == 10)].reset_index(drop=True)
df
```

```python
df['Book-Title'].values
```

```python
df['Book-Title'].value_counts()
```

```python
df['Book-Author'].values
```

```python
df['Book-Author'].value_counts()
```

```python
import re
df['Year-Of-Publication'] = df['Year-Of-Publication'].apply(lambda x: re.findall(r'\d{4}', str(x))[0] if re.findall(r'\d{4}', str(x)) else None)
df['Year-Of-Publication'] = df['Year-Of-Publication'].astype(float)
df['Year-Of-Publication'].value_counts()
```

```python
df['Description'].value_counts()
```

```python
df['Description'] = df['Description'].str.strip('"')
```

```python
df['word_count_per_desc'] = df['Description'].str.split().str.len()
```

```python
df['word_count_per_desc'].value_counts()
```

```python
df = df[df['word_count_per_desc'] > 5].reset_index(drop=True)
df.head()
```

```python
df =df.drop(columns=["word_count_per_desc"])
df.head()
```
