```python
from langchain_community.document_loaders import TextLoader
from langchain_text_splitters import CharacterTextSplitter
from langchain_chroma import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings
import pandas as pd
```

```python
books = pd.read_csv('../../data/processed/current_books.csv')
books
```

```python
books.shape
```

```python
books['Description'] = books['Description'].str.strip("")
books['Description'] = books['Description'].str.strip('"')
books['Description'] = books['Description'].str.strip("'")
books['Description'] = books['Description'].str.strip('"')
books['Description'] = books['Description'].str.strip("'")
```

```python
books["ISBN_desc"] = books['ISBN'].astype(str) + " " + books['Description'].astype(str)
```

```python
books["ISBN_desc"].to_csv('../../data/processed/ISBN_desc.txt' , sep = "\n", index = False , header=False)
loader = TextLoader('../../data/processed/ISBN_desc.txt')
raw_documents = loader.load()
text_splitter = CharacterTextSplitter(separator="\n", chunk_overlap=0)
documents = text_splitter.split_documents(raw_documents)
```

```python
documents[0]
```

```python
persist_dir = "../../data/processed/vectorstores/chroma_db"
# create / persist when building:
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")
vectorstore = Chroma.from_documents(documents, embeddings, persist_directory=persist_dir)
```

```python
query = "book about machine learning and data science"
docs = vectorstore.similarity_search(query , k = 5)
docs
```

```python
import re 
def retrieve_semantic_recommendations(
        query: str,
        top_k: int = 5,
) -> pd.DataFrame:
    recs = vectorstore.similarity_search(query, k = top_k)

    books_list = []

    for rec in recs:
        ans = ""
        for i in range(0 , rec.page_content.find(" ")): 
            if rec.page_content[i] == "'" : 
                continue
            if rec.page_content[i] == '"' : 
                continue
            ans += rec.page_content[i]
        books_list.append(ans)


    return books[books["ISBN"].isin(books_list)]
```

```python
retrieve_semantic_recommendations("A book about machine learning and data science")
```
